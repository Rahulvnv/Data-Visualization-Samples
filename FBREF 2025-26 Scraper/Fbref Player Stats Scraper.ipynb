```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, WebDriverException
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

player_url = "https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats"
service = Service(r"C:\Users\Rahul\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)
driver.get(player_url)
df=driver.find_element(By.XPATH, '//*[@id="stats_standard"]').get_attribute('innerHTML').split(sep='\n')
```

```python
df1=df
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Assuming 'data_list' is your list of strings
html_content = "".join(df1)

# 1. Use Pandas to parse the table strings directly
# We wrap it in <table> tags to ensure the parser recognizes it
tables = pd.read_html(f"<table>{html_content}</table>")
df = tables[0]

# 2. Clean up the Multi-Index Header
# The data has 'Over Headers' (Playing Time, Performance, etc.)
# This flattens them so you just have 'Player', 'Gls', 'xG', etc.
df.columns = df.columns.get_level_values(-1)

# 3. Basic Cleaning
# Remove rows that are just header repeats (common in long sports tables)
df = df[df['Player'] != 'Player']

# Display the result
df.columns.unique()
```
